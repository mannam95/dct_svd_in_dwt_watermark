```python
import os
from matplotlib import cm, pyplot as plt
import numpy as np
import config
from PIL import Image
from dct import apply_2d_dct_all_blocks, get_watermarkbits_from_dct_blocks
from dwt import dwt_2d
from helpers import create_non_overlapping_blocks, read_image
from pathlib import Path
from watermark_encryption import watermark_image_decryption, decrypt_watermark, get_watermark_img
```

```python
def return_encrypted_watermarking_bits(image_path):
    img = Image.open(image_path)

    img = np.asarray(img.convert(mode='L'))

    # img = np.asarray(img.resize((256, 256)))

    LL, (LH, HL, HH) = dwt_2d(img, 'haar') # get the coefficients of DWT transform

    non_overlapping_blocks = create_non_overlapping_blocks(LL, config.dct_block_size) # creates non-overlapping blocks

    dct_blocks = apply_2d_dct_all_blocks(non_overlapping_blocks) # applies dct to all the non-overlapping blocks

    watermark_bits = get_watermarkbits_from_dct_blocks(dct_blocks) # gets the watermark bits

    return watermark_bits
```

```python
img = read_image(config.train_images_path + "lena_512x512.gif")
watermark_img = get_watermark_img() # Gets a 32*32 binary watermark logo
with open(config.global_enc_key_path, 'rb') as f:
    enc_key = np.load(f)
```

# When alpha is 1.1

```python
image_name = "lena_512x512_11.gif"
watermark_bits = return_encrypted_watermarking_bits(config.embedded_images_path + image_name)
decrypted_watermark = decrypt_watermark(watermark_bits, enc_key)

plt.figure(figsize=(20, 5))

# Plot original watermark image
ax = plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title( "Cover image")
plt.gray()

# Plot encrypted watermark image
ax = plt.subplot(1, 4, 2)
plt.imshow(read_image(config.embedded_images_path + image_name))
plt.title( "Cover + Watermarked image")
plt.gray()

# Plot encrypted watermark image
ax = plt.subplot(1, 4, 3)
plt.imshow(watermark_img)
plt.title( "original Watermark Logo")
plt.gray()

# Plot encrypted watermark image
ax = plt.subplot(1, 4, 4)
plt.imshow(decrypted_watermark)
plt.title( "Extracted Watermark Logo")
plt.gray()

plt.show()
```
